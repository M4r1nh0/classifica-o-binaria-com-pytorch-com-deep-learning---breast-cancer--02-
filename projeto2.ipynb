```
import pandas as pd #para a base da dados
import numpy as np #pros vetores
import torch.nn as nn #pra rede neural or neural network
from skorch import NeuralNetBinaryClassifier #Uma rede neural que vai traalhar com um problema binario...
import torch
import seaborn as sns
from sklearn.model_selection import cross_val_score #
#torch.__version__
```

```
np.random.seed(123)
torch.manual_seed(123)
```

```
previsores = pd.read_csv('entradas_breast.csv')
classe = pd.read_csv('saidas_breast.csv')
```

```
previsores = np.array(previsores, dtype = 'float32')
```

```
classe = np.array(classe, dtype='float32').squeeze(1)
```

```
class classificador_torch(nn.Module): #classe que vai herda diretamente o modelo do torch
  def __init__(self):
    super().__init__()  #criando um super que vai herdar tudo 
    #30 > 16 > 16 > 1
    self.dense0 = nn.Linear(30, 16) #criando as entradas
    torch.nn.init.uniform_(self.dense0.weight) #inicializando os pesos com um estatistica uniforme
    self.activation0 = nn.ReLU() #primeira função de ativaçã
    self.dense1 = nn.Linear(16, 16) #criando a outra camada oculta
    torch.nn.init.uniform_(self.dense1.weight) #inicializando os pesos 
    self.activation1 = nn.ReLU() #segunda funçãoo de ativação
    self.dense2 = nn.Linear(16,1) #ultima camada
    torch.nn.init.uniform_(self.dense2.weight) #inicializando os pesos da ultima camada
    self.output = nn.Sigmoid()

  def forward(self, X):
    X = self.dense0(X)
    X = self.activation0(X)
    X = self.dense1(X)
    X = self.activation1(X)
    X = self.dense2(X)
    X = self.output(X)
    return X
```

```
classificador_sklearn = NeuralNetBinaryClassifier(module=classificador_torch,
                                                  criterion=torch.nn.BCELoss,
                                                  optimizer=torch.optim.Adam,
                                                  lr=0.001,
                                                  optimizer__weight_decay=0.0001,
                                                  max_epochs=100,
                                                  batch_size=10,
                                                  train_split=False)
```

```
resultados = cross_val_score(classificador_sklearn, previsores, classe, cv = 10, scoring = 'accuracy')
```

```
class classificador_torch(nn.Module): #classe que vai herda diretamente o modelo do torch
  def __init__(self):
    super().__init__()  #criando um super que vai herdar tudo 
    #30 > 16 > 16 > 1
    self.dense0 = nn.Linear(30, 16) #criando as entradas
    torch.nn.init.uniform_(self.dense0.weight) #inicializando os pesos com um estatistica uniforme
    self.activation0 = nn.ReLU() #primeira função de ativaçã
    self.drop0 = nn.Dropout(0.2) #dropout zera 25% dos neuronios
    self.dense1 = nn.Linear(16, 16) #criando a outra camada oculta
    torch.nn.init.uniform_(self.dense1.weight) #inicializando os pesos 
    self.activation1 = nn.ReLU() #segunda funçãoo de ativação
    self.drop1 = nn.Dropout(0.2)
    self.dense2 = nn.Linear(16,1) #ultima camada
    torch.nn.init.uniform_(self.dense2.weight) #inicializando os pesos da ultima camada
    self.output = nn.Sigmoid()

  def forward(self, X):
    X = self.dense0(X)
    X = self.activation0(X)
    X = self.drop0(X)
    X = self.dense1(X)
    X = self.activation1(X)
    X = self.drop1(X)
    X = self.dense2(X)
    X = self.output(X)
    return X
```

```
classificador_sklearn = NeuralNetBinaryClassifier(module=classificador_torch,
                                                  criterion=torch.nn.BCELoss,
                                                  optimizer=torch.optim.Adam,
                                                  lr=0.001,
                                                  optimizer__weight_decay=0.0001,
                                                  max_epochs=100,
                                                  batch_size=10,
                                                  train_split=False)
```
